```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
import seaborn as sn
```

# Exploratory Data Analysis

```python
bank_data = pd.read_csv("Variant II.csv")
```

```python
bank_data.head()
```

```python
fraud_bool = pd.unique(bank_data.loc[:,'fraud_bool'])
fraud_bool_counts = bank_data.loc[:,'fraud_bool'].value_counts()

plt.figure(figsize=(5,3))
plt.bar(x = fraud_bool, height = fraud_bool_counts, color = ['lightsteelblue','crimson'])
```
